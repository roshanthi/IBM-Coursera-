```python
import numpy as np

X_train = np.load('X_train_NonDL.npy')
Y_train = np.load('Y_train_NonDL.npy')
X_test = np.load('X_test_NonDL.npy')
Y_test = np.load('Y_test_NonDL.npy')
```

```python
print(np.shape(X_train))
print(np.shape(Y_train))
print(np.shape(X_test))
print(np.shape(Y_test))
```

```python
#Creating the Support Vector MAchine classifier
from sklearn import svm
import datetime as dt

C = 5
gamma = 0.05
classifier = svm.SVC(C=C,gamma=gamma)


start_time = dt.datetime.now()
print('Start learning at {}'.format(str(start_time)))
classifier.fit(X_train, Y_train)
end_time = dt.datetime.now() 
print('Stop learning {}'.format(str(end_time)))
elapsed_time= end_time - start_time
print('Elapsed learning {}'.format(str(elapsed_time)))
```

```python
score=classifier.predict(X_test)
```

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(Y_test,score)
print(accuracy)
```

```python
#Creating K-Nearest Neighbors classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score

classifier=KNeighborsClassifier()


classifier.fit(X_train, Y_train)


score=classifier.predict(X_test)
accuracy=metrics.accuracy_score(Y_test,score)
print(accuracy)
```
